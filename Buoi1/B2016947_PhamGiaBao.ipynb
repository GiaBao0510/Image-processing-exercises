```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = plt.imread('/content/drive/MyDrive/XuLyAnh/TH1/TH1/Photo00002.jpg')
plt.imshow(img1)    #Hiển thị hình ảnh
```

```python
r = img1[:,:,0]   #Lấy kênh màu đỏ vì reb ở vị trí số 0 trong (R,G,B)
plt.imshow(r,cmap='gray')   #Hiển thị hình ảnh ở dạng mức xám
```

```python
type(img1)
print(f'Shape: {img1.shape}')
print(f'ndim: {img1.ndim}')
print(f'size: {img1.size}')
print(f'dtype: {img1.dtype}')
print(f'nbytes: {img1.nbytes}')
```


- Shap dùng để hiển thị kích thước hình ảnh, dưới dạng (height,withd,channels)
- ndim: thường dùng để hiển thị số chiều của hình ảnh, ảnh mức xám thì in ra giá trị là 2 , ảnh màu thì giá trị là 3
- size: dùng để hiển thị tổng phần tử trong ảnh.
- dtype: dùng để hiển thị kiểu dữ liệu của mỗi phần tử ảnh.
- nbytes: dùng để hiển thị tổng kích thước của hình ảnh trong bộ nhớ.

```python
hist, bins = np.histogram(r.ravel(), bins=256, range=(0,256))
'''
    + ravel(): chuyển ma trân 2 chiều trên về ma trận 1 chiều
    + Hàm histogram(): dùng để đếm các phần tử của r trong khoảng range(rừ 0 đến 255), chia ra làm 256 nhóm giá trị
    , Các số lượng sẽ ghi vào mảng bins.
    + hist là vector tần suất xuất các giá trị trong từng khoảng
'''
plt.title("Red histogram")
plt.bar(bins[:-1],hist) # Vẽ biểu đồ cột chứa mảng hist và bins trong đó bin bị giới hạn bỏ đi phần tử cuối
```

```python
plt.subplots(nrows = 2, ncols=2)    #Tạo ra 1 figure có 2 hàng và 2 cột

# Hiển thị ảnh gốc
plt.subplot(2,2,1)      #(số hàng:2, số cột:2 ,vị trí xuất hiện:1)
plt.title('Original image')
plt.imshow(img1)

#Hiển thị tổ chức đồ của kênh màu đỏ
r = img1[:,:,0]             #Lấy kênh màu đỏ vì reb ở vị trí số 0 trong (R,G,B)
hist, bins = np.histogram(r.ravel(), bins=256, range=(0,256))
plt.subplot(2,2,2)      #(số hàng:2, số cột:2 ,vị trí xuất hiện:2)
plt.title("Reb histogram")
plt.bar(bins[:-1], hist)

#Hiển thị tổ chức đồ của kênh màu xanh lá
g = img1[:,:,1]             #Lấy kênh màu xanh lá vì reb ở vị trí số 1 trong (R,G,B)
hist, bins = np.histogram(g.ravel(), bins=256, range=(0,256))
plt.subplot(2,2,3)       #(số hàng:2, số cột:2 ,vị trí xuất hiện:3)
plt.title("Green histogram")
plt.bar(bins[:-1], hist)

#Hiển thị tổ chức đồ của kênh màu xanh dương
b = img1[:,:,2]         #Lấy kênh màu xanh dương vì reb ở vị trí số 2 trong (R,G,B)
hist, bins = np.histogram(b.ravel(), bins=256, range=(0,256))
plt.subplot(2,2,4)       #(số hàng:2, số cột:2 ,vị trí xuất hiện:4)
plt.title("Blue histogram")
plt.bar(bins[:-1], hist)
```
