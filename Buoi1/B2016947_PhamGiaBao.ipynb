```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = plt.imread('/content/drive/MyDrive/XuLyAnh/TH1/TH1/Photo00002.jpg')
plt.imshow(img1)    #Hin th hnh nh
```

```python
r = img1[:,:,0]   #Ly knh mu  v reb  v tr s 0 trong (R,G,B)
plt.imshow(r,cmap='gray')   #Hin th hnh nh  dng mc xm
```

```python
type(img1)
print(f'Shape: {img1.shape}')
print(f'ndim: {img1.ndim}')
print(f'size: {img1.size}')
print(f'dtype: {img1.dtype}')
print(f'nbytes: {img1.nbytes}')
```


- Shap dng  hin th kch thc hnh nh, di dng (height,withd,channels)
- ndim: thng dng  hin th s chiu ca hnh nh, nh mc xm th in ra gi tr l 2 , nh mu th gi tr l 3
- size: dng  hin th tng phn t trong nh.
- dtype: dng  hin th kiu d liu ca mi phn t nh.
- nbytes: dng  hin th tng kch thc ca hnh nh trong b nh.

```python
hist, bins = np.histogram(r.ravel(), bins=256, range=(0,256))
'''
    + ravel(): chuyn ma trn 2 chiu trn v ma trn 1 chiu
    + Hm histogram(): dng  m cc phn t ca r trong khong range(r 0 n 255), chia ra lm 256 nhm gi tr
    , Cc s lng s ghi vo mng bins.
    + hist l vector tn sut xut cc gi tr trong tng khong
'''
plt.title("Red histogram")
plt.bar(bins[:-1],hist) # V biu  ct cha mng hist v bins trong  bin b gii hn b i phn t cui
```

```python
plt.subplots(nrows = 2, ncols=2)    #To ra 1 figure c 2 hng v 2 ct

# Hin th nh gc
plt.subplot(2,2,1)      #(s hng:2, s ct:2 ,v tr xut hin:1)
plt.title('Original image')
plt.imshow(img1)

#Hin th t chc  ca knh mu 
r = img1[:,:,0]             #Ly knh mu  v reb  v tr s 0 trong (R,G,B)
hist, bins = np.histogram(r.ravel(), bins=256, range=(0,256))
plt.subplot(2,2,2)      #(s hng:2, s ct:2 ,v tr xut hin:2)
plt.title("Reb histogram")
plt.bar(bins[:-1], hist)

#Hin th t chc  ca knh mu xanh l
g = img1[:,:,1]             #Ly knh mu xanh l v reb  v tr s 1 trong (R,G,B)
hist, bins = np.histogram(g.ravel(), bins=256, range=(0,256))
plt.subplot(2,2,3)       #(s hng:2, s ct:2 ,v tr xut hin:3)
plt.title("Green histogram")
plt.bar(bins[:-1], hist)

#Hin th t chc  ca knh mu xanh dng
b = img1[:,:,2]         #Ly knh mu xanh dng v reb  v tr s 2 trong (R,G,B)
hist, bins = np.histogram(b.ravel(), bins=256, range=(0,256))
plt.subplot(2,2,4)       #(s hng:2, s ct:2 ,v tr xut hin:4)
plt.title("Blue histogram")
plt.bar(bins[:-1], hist)
```
